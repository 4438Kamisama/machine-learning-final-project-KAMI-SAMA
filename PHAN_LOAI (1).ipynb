## 11. Xây dựng pipeline tiền xử lý

## Xây dựng tiền xử lý dữ liệu

Để dữ liệu đầu vào phù hợp với các thuật toán học máy, bài làm xây dựng quy trình tiền xử lý riêng cho từng nhóm biến.

Đối với các biến dạng số như tuổi, khoảng cách bay, thời gian trễ và các điểm đánh giá dịch vụ, giá trị thiếu được xử lý và dữ liệu được chuẩn hóa bằng StandardScaler. Việc chuẩn hóa giúp các biến số có cùng thang đo, đặc biệt hữu ích với các mô hình như Logistic Regression, KNN và Naive Bayes.

Đối với các biến dạng phân loại như giới tính, loại khách hàng, loại chuyến bay và hạng vé, dữ liệu chữ được mã hóa bằng One-Hot Encoding. Cách mã hóa này giúp chuyển các giá trị dạng chữ thành dạng số để mô hình có thể xử lý.

Toàn bộ quá trình tiền xử lý được đưa vào Pipeline để đảm bảo dữ liệu train, test và dữ liệu dự đoán mới đều được xử lý thống nhất. Điều này giúp giảm lỗi trong quá trình huấn luyện và tránh sai lệch khi đánh giá mô hình.

```python
numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

try:
    categorical_transformer = Pipeline(steps=[
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("onehot", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
    ])
except TypeError:
    categorical_transformer = Pipeline(steps=[
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("onehot", OneHotEncoder(handle_unknown="ignore", sparse=False))
    ])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features)
    ]
)
```

###Nhận xét:
Pipeline giúp xử lý dữ liệu thống nhất và tránh rò rỉ dữ liệu. Các bước tiền xử lý chỉ học trên tập train rồi áp dụng cho tập test.

## 12. Hàm đánh giá mô hình

```python
def evaluate_model(model, X_test, y_test, model_name):
    y_pred = model.predict(X_test)

    if hasattr(model.named_steps["model"], "predict_proba"):
        y_proba = model.predict_proba(X_test)[:, 1]
        roc_auc = roc_auc_score(y_test, y_proba)
    else:
        y_proba = None
        roc_auc = np.nan

    result = {
        "Model": model_name,
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred, zero_division=0),
        "Recall": recall_score(y_test, y_pred, zero_division=0),
        "F1-score": f1_score(y_test, y_pred, zero_division=0),
        "ROC-AUC": roc_auc
    }

    print("Mô hình:", model_name)
    print("Accuracy :", round(result["Accuracy"], 4))
    print("Precision:", round(result["Precision"], 4))
    print("Recall   :", round(result["Recall"], 4))
    print("F1-score :", round(result["F1-score"], 4))
    print("ROC-AUC  :", round(result["ROC-AUC"], 4))

    return result, y_pred, y_proba
```

###Nhận xét:
Hàm này giúp đánh giá các mô hình theo cùng một cách. Các chỉ số được sử dụng gồm Accuracy, Precision, Recall, F1-score và ROC-AUC.

Sau khi so sánh các mô hình, Random Forest là mô hình có kết quả tốt nhất. Trên tập test, mô hình đạt Accuracy = 0.9631, Precision = 0.9704, Recall = 0.9437, F1-score = 0.9569 và ROC-AUC = 0.9941. Đây là kết quả cao, cho thấy mô hình có khả năng phân loại tốt giữa nhóm hành khách hài lòng và nhóm không hài lòng/trung lập.

Kết quả cũng cho thấy các mô hình ensemble như Random Forest, Random Forest Tuned và Extra Trees hoạt động tốt hơn các mô hình đơn giản hơn như Logistic Regression, KNN và Gaussian Naive Bayes. Điều này có thể do dữ liệu chứa nhiều mối quan hệ phi tuyến giữa các đặc trưng về dịch vụ và mức độ hài lòng của hành khách.

Feature Importance cho thấy các yếu tố có ảnh hưởng lớn đến dự đoán gồm chất lượng làm thủ tục trực tuyến, dịch vụ wifi trên máy bay, hạng vé, loại chuyến bay, giải trí trên máy bay, độ thoải mái của ghế ngồi và khoảng cách chuyến bay. Đây đều là các yếu tố hợp lý vì chúng có liên quan trực tiếp đến trải nghiệm của hành khách.

Nhìn chung, mô hình Random Forest có thể được sử dụng để hỗ trợ dự đoán mức độ hài lòng của hành khách, đồng thời giúp doanh nghiệp hàng không nhận diện những yếu tố dịch vụ quan trọng cần cải thiện. Tuy nhiên, kết quả của mô hình chỉ mang tính tham khảo trong phạm vi học tập và phân tích dữ liệu, không thay thế hoàn toàn cho khảo sát thực tế hoặc đánh giá chuyên sâu từ doanh nghiệp.
